```python
from pydantic import BaseModel, ConfigDict, ValidationError
```

```python
class Tags(BaseModel):
    tags: str = "misc"

class Supplier(BaseModel):
    supplier: str
    contact_email: str = "TBD"

# "None" berücksichtigen: https://pydantic.dev/docs/validation/latest/concepts/models/#nested-models
class Product(BaseModel):
    model_config = ConfigDict(strict=True)

    id: int
    name: str
    price: float
    is_offer: bool
    tags: Tags 
    supplier: Supplier
    
```
